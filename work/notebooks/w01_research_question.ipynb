## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

To validate this lane, I analyzed the starter dataset content_refresh_anonymized.csv. Based on the code execution below, we found the following real numbers from the inventory:


* Total eligible verified pages: 29,998
* Total pages currently in a downward trend: 16,261
* Percentage of inventory facing decline risk: 54.21%

This means more than half of the site's content is actively losing performance, which strongly proves that building a learned prioritization queue for content refresh is worth our next 7 weeks.  

```python
import pandas as pd
import numpy as np

try:
    df = pd.read_csv('content_refresh_anonymized.csv', on_bad_lines='skip', low_memory=False)
except FileNotFoundError:
    df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv', on_bad_lines='skip', low_memory=False)

required_cols = ['content_age_days', 'impressions_90d', 'content_id', 'trend_direction']
if all(col in df.columns for col in required_cols):
    df_clean = df[(df['content_age_days'] >= 90) & (df['impressions_90d'] > 0)].drop_duplicates(subset=['content_id'])

    total_pages = len(df_clean)
    declining_pages = len(df_clean[df_clean['trend_direction'] == 'down'])
    pct_declining = (declining_pages / total_pages) * 100

    print(f"--- Real Numbers from Starter Dataset ---")
    print(f"1. Total eligible verified pages: {total_pages:,}")
    print(f"2. Total pages currently in a downward trend: {declining_pages:,}")
    print(f"3. Percentage of inventory facing decline risk: {pct_declining:.2f}%")
else:
    print(" Missing required columns in the loaded dataset block ")
```
